```python
df.fillna(0,inplace=True)
print(df.to_string())
```

```python
Q1 = np.percentile(df['math score'], 25,
                   interpolation = 'midpoint')
 
Q3 = np.percentile(df['math score'], 75,
                   interpolation = 'midpoint')
IQR = Q3 - Q1
IQR
```

```python
upper = df['math score'] >= (Q3+1.5*IQR)
 
print("Upper bound:",upper)
print(np.where(upper))
 
# Below Lower bound
lower = df['math score'] <= (Q1-1.5*IQR)
print("Lower bound:", lower)
print(np.where(lower))
```

```python
Q1 = np.percentile(df['math score'], 25,
                   interpolation = 'midpoint')
 
Q3 = np.percentile(df['math score'], 75,
                   interpolation = 'midpoint')
IQR = Q3 - Q1
upper = np.where(df['math score'] >= (Q3+1.5*IQR))
 
# Below Lower bound
lower = np.where(df['math score'] <= (Q1-1.5*IQR))
df.drop(upper[0], inplace = True)
df.drop(lower[0], inplace = True)
```
