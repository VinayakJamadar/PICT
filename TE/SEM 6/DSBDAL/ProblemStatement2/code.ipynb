# 4. Turn categorical variables into quantitative variables in Python

```python
# Turn categorical variables into quantitative variables in Python
# Here categorical variables 'Gender' is turned to quantitative variables 'female' and 'male' using get_dummies

# Here we consider different dataset as follows
df = pd.read_csv('data1.csv')
df_gender = pd.get_dummies(df['Gender'])
df_new = pd.concat([df, df_gender], axis=1)
print(df_new)
```
