```python
import pandas as pd
import numpy as np

df = pd.read_csv('iris.csv')
df
```

```python
df['Species'].unique()
```

# Group by qualitative (categorical) variable - i.e Group by 'Species

Percentile - 25%, 50% and 75%

```python
# default quantile at 0.5 or 50%
df.groupby(['Species']).quantile()
```

```python
# quantile at 0.25 or 25%
df.groupby(['Species']).quantile(0.25)
```

```python
# quantile at 0.75 or 75%
df.groupby(['Species']).quantile(0.75)
```

Mean

```python
df.groupby(['Species']).mean()
```

Standard Deviation

```python
df.groupby(['Species']).std()
```

Min

```python
df.groupby(['Species']).min()
```

Max

```python
df.groupby(['Species']).max()
```

Median

```python
df.groupby(['Species']).median()
```
