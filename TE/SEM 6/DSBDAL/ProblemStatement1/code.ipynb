# 4. Data Formatting and Data Normalization

```python
# Convert 'YearsExperience' column data type from float64 to int64
df['YearsExperience'] = df['YearsExperience'].astype(dtype = 'int64')
df
```

Data Normalization

```python
# Filling NULL or NAN values of 'Salary' column with 'mean'
df['Salary'].fillna(value=df['Salary'].mean(), inplace=True)
df = df.astype({'Salary': 'int64'})
df
```

```python
df1 = df.copy()
# Z-Score Standardization
df1['Salary'] = (df1['Salary'] - df1['Salary'].mean()) / df1['Salary'].std()
df1
```

```python
df2 = df.copy()
# Min-Max Scaling (Normalization)
df2['Salary'] = (df2['Salary'] - df2['Salary'].min()) / (df2['Salary'].max() - df2['Salary'].min())
df2
```
